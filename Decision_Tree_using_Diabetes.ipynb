```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("")
print(df.head())
```

```python
#separate features and labels separately first
# x->features; y->labels


x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#training with x and y test data
from sklearn.tree import DecisionTreeClassifier
# parameters are gini,entropy,information gain. These are used for improve the model efficiency
# dt=DecisionTreeClassifier(criterion='entropy') # can give parameter or not
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
#testthe learning of model with test data
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))


# everyone gets different accuracy values because of no preprocessing and encoding
# what is enconding: model doesn't support categorical value(-> strings) so, we change that into numerical data
# what is Standardization : the units and range are different, so getting them all in same range
```
