```python
# top 5 sales on that one day by amount

filtered_data=dt[dt['DATE']=='01.10.23']
filtered_data.groupby('INVOICE_NO')
sorted_data=dt.sort_values(by='AMOUNT',ascending=False)
print(sorted_data.head())
```

```python
data=pd.read_csv("/content/Data.csv")
```

```python
fdata=data.groupby('Date')['Final Total'].mean().reset_index(name='avg')
df=fdata.sort_values(by='avg',ascending=False).head(1)
print(df)
```

```python
# it returns as dataframe
fdata=data.groupby('Invoice No.').count()
print(fdata)

#it return as value
data['Invoice No.'].nunique()

```

```python
fdata=data[(data['Date']>='2023-09-01') & (data['Date']<='2023-09-31')]
fdata
df=fdata.groupby('Item Name')['Qty.'].sum()
df.sort_values(ascending=False).head()
```

```python
fdata=data[(data['Date']>='2023-07-01') & (data['Date']<='2023-07-31')]
df=fdata.groupby('Category')['Qty.'].sum().sort_values(ascending=True).head()
print(df)
```

```python
df =data.groupby('Category')['Qty.'].sum()
plt.pie(df,labels=df.index)
plt.show()
```
